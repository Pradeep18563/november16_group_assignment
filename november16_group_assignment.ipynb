# Compare churn count with respect to gender.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel(r"C:\Users\HP\OneDrive\Desktop\Telco-Customer-Churn.xlsx")
```

```python
# Read and count
df=pd.DataFrame(data)
a=df.value_counts('gender')
print(a)
```

```python
df['gender'].value_counts().plot(kind='bar')
plt.xlabel('gender')       
plt.hist(['Female','Male'],bins=[60,70,80,100,120],label=['Female','Male'])
plt.grid(True)
plt.ylim((500,5000))
plt.legend()
```
